West Bengal AITC

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

url = "https://results.eci.gov.in/PcResultGenJune2024/partywisewinresult-140S25.htm"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table', {'class': 'table table-striped table-bordered'})

data = []
for row in table.find_all('tr')[1:]:
    cols = row.find_all('td')
    data.append([col.text.strip() for col in cols])

df = pd.DataFrame(data, columns=['S.No', 'Parliament Constituency', 'Winning Candidate', 'Total Votes', 'Margin'])

# Create an Excel sheet
df.to_excel('partywisewinresult.xlsx', index=False)

# Create a pie chart using matplotlib
plt.pie(df['Total Votes'], labels=df['Parliament Constituency'], autopct='%1.1f%%')
plt.title('Party-wise Winning Results')
plt.show()
```
