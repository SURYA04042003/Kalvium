```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# Send a GET request to the ECI website
url = "https://results.eci.gov.in/PcResultGenJune2024/partywisewinresult-369S03.htm"
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Extract the table data from the HTML content
table = soup.find('table', {'class': 'table table-striped table-bordered'})
rows = table.find_all('tr')

# Create a list to store the data
data = []

# Iterate over the rows and extract the data
for row in rows[1:]:
    cols = row.find_all('td')
    data.append([col.text.strip() for col in cols])

# Create a Pandas DataFrame from the data
df = pd.DataFrame(data, columns=['S.No', 'Parliament Constituency', 'Winning Candidate', 'Total Votes', 'Margin'])

# Generate a bar graph of the total votes
plt.figure(figsize=(10, 6))
plt.bar(df['Parliament Constituency'], df['Total Votes'])
plt.xlabel('Parliament Constituency')
plt.ylabel('Total Votes')
plt.title('Total Votes by Parliament Constituency')
plt.show()

# Generate a bar graph of the margin
plt.figure(figsize=(10, 6))
plt.bar(df['Parliament Constituency'], df['Margin'])
plt.xlabel('Parliament Constituency')
plt.ylabel('Margin')
plt.title('Margin by Parliament Constituency')
plt.show()


df.to_excel('eci_data.xlsx', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# Send a GET request to the ECI website
url = "https://results.eci.gov.in/PcResultGenJune2024/partywisewinresult-369S03.htm"
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Extract the table data from the HTML content
table = soup.find('table', {'class': 'table table-striped table-bordered'})
rows = table.find_all('tr')

# Create a list to store the data
data = []

# Iterate over the rows and extract the data
for row in rows[1:]:
    cols = row.find_all('td')
    data.append([col.text.strip() for col in cols])

# Create a Pandas DataFrame from the data
df = pd.DataFrame(data, columns=['S.No', 'Parliament Constituency', 'Winning Candidate', 'Total Votes', 'Margin'])

# Count the number of seats won by each party
party_counts = df['Parliament Constituency'].value_counts()

# Generate a bar graph of the total votes
plt.figure(figsize=(10, 6))
plt.bar(df['Parliament Constituency'].unique(), df.groupby('Parliament Constituency')['Total Votes'].sum())
plt.xlabel('Parliament Constituency')
plt.ylabel('Total Votes')
plt.title('Total Votes by Parliament Constituency')
plt.show()

# Generate a bar graph of the margin
plt.figure(figsize=(10, 6))
plt.bar(df['Parliament Constituency'].unique(), df.groupby('Parliament Constituency')['Margin'].sum())
plt.xlabel('Parliament Constituency')
plt.ylabel('Margin')
plt.title('Margin by Parliament Constituency')
plt.show()

# Generate a pie chart of the number of seats won by each party
plt.figure(figsize=(8, 8))
plt.pie(party_counts, labels=party_counts.index, autopct='%1.1f%%')
plt.title('Party-wise Winning Seats')
plt.show()

# Save the data to an Excel sheet
df.to_excel('eci_data_369S03.xlsx', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# Send a GET request to the ECI website
url = "https://results.eci.gov.in/PcResultGenJune2024/partywisewinresult-1998S03.htm"
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Extract the table data from the HTML content
table = soup.find('table', {'class': 'table table-striped table-bordered'})
rows = table.find_all('tr')

# Create a list to store the data
data = []

# Iterate over the rows and extract the data
for row in rows[1:]:
    cols = row.find_all('td')
    data.append([col.text.strip() for col in cols])

# Create a Pandas DataFrame from the data
df = pd.DataFrame(data, columns=['S.No', 'Party', 'Won', 'Leading', 'Total'])

# Count the number of seats won by each party
party_counts = df['Party'].value_counts()

# Generate a bar graph of the number of seats won by each party
plt.figure(figsize=(10, 6))
plt.bar(party_counts.index, party_counts.values)
plt.xlabel('Party')
plt.ylabel('Number of Seats Won')
plt.title('Party-wise Winning Seats')
plt.show()

# Generate a pie chart of the number of seats won by each party
plt.figure(figsize=(8, 8))
plt.pie(party_counts.values, labels=party_counts.index, autopct='%1.1f%%')
plt.title('Party-wise Winning Seats')
plt.show()

# Save the data to an Excel sheet
df.to_excel('eci_data_1998S03.xlsx', index=False)
```
