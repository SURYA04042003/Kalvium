```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = '/content/election_results.xlsx'
df = pd.read_excel(file_path)

# Display the DataFrame
print(df)

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a bar chart for the total seats won by each party
plt.bar(df['Party'], df['Total'], color=df['Color'])

# Add labels and title
plt.xlabel('Party')
plt.ylabel('Total Seats Won')
plt.title('Total Seats Won by Each Party in Assam (2024)')
plt.xticks(rotation=45)

# Display the bar chart
plt.show()

# Set the figure size for the pie chart
plt.figure(figsize=(10, 6))

# Create a pie chart for the vote share of each party
plt.pie(df['Vote Share (%)'], labels=df['Party'], colors=df['Color'], autopct='%1.1f%%', startangle=140)

# Add title
plt.title('Vote Share of Each Party in Assam (2024)')

# Display the pie chart
plt.show()
```
