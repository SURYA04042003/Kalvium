Kerala INC

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = '/content/election_winning_candidates_kerala.xlsx'
df_winning_candidates = pd.read_excel(file_path, sheet_name='Winning Candidates')

# Display the data
print("Winning Candidates DataFrame:")
print(df_winning_candidates)

# Set Seaborn style
sns.set(style="whitegrid")

# Bar Plot for Total Votes per Constituency
plt.figure(figsize=(12, 8))
sns.barplot(data=df_winning_candidates, x='Parliament Constituency', y='Total Votes', palette='Blues_d')
plt.xlabel('Parliament Constituency')
plt.ylabel('Total Votes')
plt.title('Total Votes per Constituency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bar Plot for Margin of Victory per Constituency
plt.figure(figsize=(12, 8))
sns.barplot(data=df_winning_candidates, x='Parliament Constituency', y='Margin', palette='Oranges_d')
plt.xlabel('Parliament Constituency')
plt.ylabel('Margin of Victory')
plt.title('Margin of Victory per Constituency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Line Plot for Total Votes and Margin of Victory per Constituency
plt.figure(figsize=(12, 8))
sns.lineplot(data=df_winning_candidates, x='Parliament Constituency', y='Total Votes', marker='o', label='Total Votes', color='b')
sns.lineplot(data=df_winning_candidates, x='Parliament Constituency', y='Margin', marker='o', label='Margin of Victory', color='r')
plt.xlabel('Parliament Constituency')
plt.ylabel('Count')
plt.title('Total Votes and Margin of Victory per Constituency')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Scatter Plot for Total Votes vs. Margin of Victory
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_winning_candidates, x='Total Votes', y='Margin', hue='Parliament Constituency', palette='viridis', s=100)
plt.xlabel('Total Votes')
plt.ylabel('Margin of Victory')
plt.title('Total Votes vs. Margin of Victory')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Kerala IUML

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://results.eci.gov.in/PcResultGenJune2024/partywisewinresult-772S11.htm"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table', {'class': 'table table-striped table-bordered'})

data = []
for row in table.find_all('tr')[1:]:
    cols = row.find_all('td')
    data.append([col.text.strip() for col in cols])

df = pd.DataFrame(data, columns=['S.No', 'Parliament Constituency', 'Winning Candidate', 'Total Votes', 'Margin'])

# Create an Excel sheet
df.to_excel('partywisewinresult.xlsx', index=False)

# Create a graph using matplotlib
import matplotlib.pyplot as plt

plt.bar(df['Parliament Constituency'], df['Total Votes'])
plt.xlabel('Parliament Constituency')
plt.ylabel('Total Votes')
plt.title('Party-wise Winning Results')
plt.show()
```
