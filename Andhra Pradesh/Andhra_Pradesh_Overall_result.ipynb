```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for Election Results for Andhra Pradesh
election_results_data = {
    "Party": ["Telugu Desam - TDP", "Yuvajana Sramika Rythu Congress Party - YSRCP",
              "Bharatiya Janata Party - BJP", "Janasena Party - JnP", "Total"],
    "Won": [16, 4, 3, 2, 25],
    "Leading": [0, 0, 0, 0, 0],
    "Total": [16, 4, 3, 2, 25]
}

# Data for Party Wise Vote Share
vote_share_data = {
    "Party": ["AIFB", "BJP", "BSP", "CPI", "CPI(M)", "CPI(ML)(L)", "INC", "NOTA",
              "SP", "TDP", "YSRCP", "Others"],
    "Vote Share (%)": [0.04, 11.28, 0.67, 0.03, 0.36, 0.01, 2.66, 1.20, 0.02, 37.79, 39.61, 6.33],
    "Votes": [13620, 3750687, 223176, 8317, 119016, 2816, 886165, 398777, 5596, 12569179, 13174874, 2106282]
}

# Create dataframes
election_results_df = pd.DataFrame(election_results_data)
vote_share_df = pd.DataFrame(vote_share_data)

# Bar graph for election results
plt.figure(figsize=(10, 6))
plt.bar(election_results_df["Party"], election_results_df["Won"], color=['#204795', '#eb3734', '#ff944d', '#9610ED', '#7CFC00'])
plt.xlabel('Party')
plt.ylabel('Seats Won')
plt.title('Election Results for Andhra Pradesh')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Pie chart for party-wise vote share
plt.figure(figsize=(10, 6))
plt.pie(vote_share_df["Votes"], labels=vote_share_df["Party"], autopct='%1.1f%%', startangle=140, colors=['#ff3333','#ff944d','#000078','#e70d21','#FF1D15','#7A68C5','#19AAED','#ff6384','#ff0000','#204795','#eb3734','#b3b3b3'])
plt.axis('equal')
plt.title('Party Wise Vote Share in Andhra Pradesh')
plt.show()
```
