```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# Send a GET request to the ECI website
url = "https://results.eci.gov.in/PcResultGenJune2024/partywisewinresult-1745S01.htm"
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Extract the table data from the HTML content
table = soup.find('table', {'class': 'table table-striped table-bordered'})
rows = table.find_all('tr')

# Create a list to store the data
data = []

# Iterate over the rows and extract the data
for row in rows[1:]:
    cols = row.find_all('td')
    data.append([col.text.strip() for col in cols])

# Create a Pandas DataFrame from the data
df = pd.DataFrame(data, columns=['S.No', 'Party', 'Won', 'Leading', 'Total'])

# Count the number of seats won by each party
party_counts = df['Party'].value_counts()

# Generate a bar graph of the number of seats won by each party
plt.figure(figsize=(10, 6))
plt.bar(party_counts.index, party_counts.values)
plt.xlabel('Party')
plt.ylabel('Number of Seats Won')
plt.title('Party-wise Winning Seats')
plt.show()

# Generate a pie chart of the number of seats won by each party
plt.figure(figsize=(8, 8))
plt.pie(party_counts.values, labels=party_counts.index, autopct='%1.1f%%')
plt.title('Party-wise Winning Seats')
plt.show()

# Save the data to an Excel sheet
df.to_excel('eci_data_1745S01.xlsx', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import re

# Send an HTTP request to the URL
url = "https://results.eci.gov.in/PcResultGenJune2024/partywisewinresult-860S01.htm"
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table with the party-wise winning results
table = soup.find('table', {'class': 'table-striped'})

# Extract the table headers
headers = [th.text.strip() for th in table.find('tr').find_all('th')]

# Extract the table data
data = []
for row in table.find_all('tr')[1:]:
    cols = row.find_all('td')
    data.append([col.text.strip() for col in cols])

# Create a Pandas DataFrame from the extracted data
df = pd.DataFrame(data, columns=headers)

# Extract the numeric part of the 'Seats Won' column using a regular expression
df.iloc[:, 1] = df.iloc[:, 1].apply(lambda x: int(re.findall("\d+", x)[0]))

# Plot the party-wise winning results
plt.figure(figsize=(10, 6))
plt.bar(df.iloc[:, 0], df.iloc[:, 1], color='#3498db')  # Blue color
plt.xlabel('Party')
plt.ylabel('Seats Won')
plt.title('Party-wise Winning Results')
plt.xticks(rotation=90)  # Rotate the x-axis labels for better readability
plt.yticks(range(0, max(df.iloc[:, 1]) + 10, 10))  # Customize the y-axis ticks
plt.legend(['Seats Won'])  # Add a legend
plt.grid(axis='y')  # Add a grid to the y-axis
plt.show()
```
