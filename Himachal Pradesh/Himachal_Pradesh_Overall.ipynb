```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = '/content/election_results (1).xlsx'
df_party_results = pd.read_excel(file_path, sheet_name='Party Results')
df_vote_share = pd.read_excel(file_path, sheet_name='Vote Share')

# Display the data
print("Party Results DataFrame:")
print(df_party_results)
print("\nVote Share DataFrame:")
print(df_vote_share)

# Plotting the Party Results
plt.figure(figsize=(8, 6))
plt.bar(df_party_results['Party'], df_party_results['Total'], color='#ff944d')
plt.xlabel('Party')
plt.ylabel('Total Seats')
plt.title('Party Results')
plt.show()

# Plotting the Vote Share
plt.figure(figsize=(10, 8))
plt.pie(df_vote_share['Votes'], labels=df_vote_share['Party'], autopct='%1.1f%%', colors=['#ff944d','#000078','#19AAED','#ff6384','#b3b3b3'])
plt.title('Party Wise Vote Share')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
