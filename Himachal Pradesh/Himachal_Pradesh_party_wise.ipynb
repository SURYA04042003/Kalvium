```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = '/content/election_winning_candidates.xlsx'
df_winning_candidates = pd.read_excel(file_path, sheet_name='Winning Candidates')

# Display the data
print("Winning Candidates DataFrame:")
print(df_winning_candidates)

# Set Seaborn style
sns.set(style="whitegrid")

# Line Plot for Total Votes per Constituency
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_winning_candidates, x='Parliament Constituency', y='Total Votes', marker='o', color='b')
plt.xlabel('Parliament Constituency')
plt.ylabel('Total Votes')
plt.title('Total Votes per Constituency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Box Plot for Total Votes
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_winning_candidates, y='Total Votes', color='cyan')
plt.ylabel('Total Votes')
plt.title('Distribution of Total Votes')
plt.tight_layout()
plt.show()

# Bar Plot for Margin of Victory per Constituency using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=df_winning_candidates, x='Parliament Constituency', y='Margin', palette='viridis')
plt.xlabel('Parliament Constituency')
plt.ylabel('Margin of Victory')
plt.title('Margin of Victory per Constituency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Pairplot for Total Votes and Margin
sns.pairplot(df_winning_candidates[['Total Votes', 'Margin']], kind='scatter', diag_kind='kde')
plt.show()
```
